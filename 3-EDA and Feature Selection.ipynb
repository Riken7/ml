## Import Dataset & Analyse

```python
import pandas as pd

# Import the dataset
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", 
           "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width", 
           "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size", 
           "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", 
           "city-mpg", "highway-mpg", "price"]

df = pd.read_csv(path, names=headers, na_values=["?"])
df.head(5)
```

```python
# Displaying the Data Columns with their types and Non-Null Count
df.info()
```

```python
# Displaying Basic Statistics of all the Attributes/Columns
df.describe(include="all")
```

```python
# Checking for the Null Values
df.isnull().sum()
```

## Step-2: Data Cleaning & Handling Missing Values

```python
# Checking the Attributes with the Outliers

import numpy as np

outliers = []

for feature in df.select_dtypes(include=np.number).columns:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    if df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].any(axis=None):
        outliers.append(feature)

print("Attributes with outliers:", outliers)
```

### Handling Missing Values/Outliers with different Methods:

- **'normalized-losses'**: With 41 missing values, the mean is used to fill these, given the relatively larger proportion.
- **'num-of-doors'**: Only 2 missing values, so mode is used to fill this categorical column.
- **'bore'**: Backward fill is applied since it might follow a trend.
- **'stroke'**: Mean is used for imputation as the number of missing values is small.
- **'horsepower'**: Median is used due to potential skewness in the data.
- **'peak-rpm'**: Mean is used due to the small proportion of missing values.
- **'price'**: Median is chosen as it is likely skewed and is an important target variable.


```python
# Handling Missing Values
# 1. Replace missing values in 'normalized-losses' (41 missing) with the mean, since it's a significant proportion of the dataset.
df["normalized-losses"] = df["normalized-losses"].fillna(df["normalized-losses"].mean())

# 2. Replace missing values in 'num-of-doors' (2 missing) with the mode, as it's a categorical variable and only 2 missing values.
df["num-of-doors"] = df["num-of-doors"].fillna(df["num-of-doors"].mode()[0])

# 3. Replace missing values in 'bore' (4 missing) using backward fill, since it may have an increasing trend or sequence.
df["bore"] = df["bore"].bfill()

# 4. Replace missing values in 'stroke' (4 missing) using the mean, assuming it's normally distributed and the missing proportion is small.
df["stroke"] = df["stroke"].fillna(df["stroke"].mean())

# 5. Replace missing values in 'horsepower' (2 missing) using the median, as the data might be skewed.
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].median())

# 6. Replace missing values in 'peak-rpm' (2 missing) using the mean, since the missing proportion is very small.
df["peak-rpm"] = df["peak-rpm"].fillna(df["peak-rpm"].mean())

# 7. Replace missing values in 'price' (4 missing) using the median, as it's likely skewed and is a crucial variable.
df["price"] = df["price"].fillna(df["price"].median())

# Verifying that no missing values remain
print("Missing values after handling:\n", df.isnull().sum())
```

## Differentiating Numerical and Categorical Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bifurcate the data into numerical and categorical attributes
numerical_features = df.select_dtypes(include=np.number).columns.tolist()
categorical_features = df.select_dtypes(include='object').columns.tolist()

print("Numerical Attributes:", numerical_features)
print("Categorical Attributes:", categorical_features)
```

## Regplot

```python
# Draw regplot between each continuous attribute and price
for feature in numerical_features:
    if feature != 'price':  # Skip the price column itself
        plt.figure(figsize=(8, 5))
        sns.regplot(x=df['price'], y=df[feature], scatter_kws={'alpha':0.5})
        plt.title(f'Regplot of {feature} vs. Price')
        plt.show()
```

These attributes show a significant relationship with the price, indicating that they are important for predicting the price of the vehicles.


## Boxplot

```python
# Draw boxplot between each categorical attribute and price
for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[feature], y=df['price'])
    plt.title(f'Boxplot of {feature} vs. Price')
    plt.show()
```
